# popularity based recommender system

```python
ratings.head()
```

```python
books.head()
```

```python
rating_with_name = books.merge(ratings,on='ISBN')
```

```python
 rating_with_name.groupby('Book-Title').count()
```

```python
rating_with_name.groupby('Book-Title').count()['Book-Rating']
```

```python
rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating = rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating.rename(columns={'Book-Rating':'num_rating'},inplace=True)
```

```python
num_rating
```

```python
avg_rating = rating_with_name.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating
```

```python
popular_df = num_rating.merge(avg_rating,on='Book-Title')
```

```python
popular_df
```

```python
popular_df = popular_df[popular_df['num_rating']>250].sort_values('avg_rating',ascending=False).head(50)
```
